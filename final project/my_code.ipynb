```python
import pandas as pd
from datetime import date
```

```python
df = pd.read_csv('./marketing_campaign.csv', sep='\t')

# rename columns
df = df.rename(columns={'Year_Birth': 'Age',
                        'Education': 'Edu',          # ['Graduation' 'PhD' 'Master' 'Basic' '2n Cycle']
                        'Marital_Status': 'Marital',
                        # ['Single' 'Together' 'Married' 'Divorced' 'Widow (寡婦)' 'Alone' 'Absurd(荒誕)' 'YOLO']
                        'Kidhome': 'kids',
                        'Teenhome': 'Teens',
                        'Dt_Customer': 'Dt_enroll',  # days from 
                        'MntWines': 'Wines',
                        'MntFruits':'Fruits',
                        'MntMeatProducts':'Meat',
                        'MntFishProducts':'Fish',
                        'MntSweetProducts':'Sweets',
                        'MntGoldProds':'Gold',
                        'NumDealsPurchases':'Deals', # Number of purchases made with a discount
                        'NumWebPurchases':'Web',
                        'NumCatalogPurchases':'Catalog',
                        'NumStorePurchases':'Store',
                        'NumWebVisitsMonth':'WebVis'})

# df['Complain']: 0 or 1
df['Age'] = 2021 - df['Age']

# sum all AcceptedCmp (range: 0 to 5)
df['Cmp'] = df['AcceptedCmp1']\
            + df['AcceptedCmp2']\
            + df['AcceptedCmp3']\
            + df['AcceptedCmp4']\
            + df['AcceptedCmp5']\
            + df['Response']      # 'Response' == 'AcceptedCmp6'

# delete useless columns
df = df.drop(columns=['ID',
                      'AcceptedCmp1',
                      'AcceptedCmp2',
                      'AcceptedCmp3',
                      'AcceptedCmp4',
                      'AcceptedCmp5',
                      'Response',      # 'Response' == 'AcceptedCmp6'
                      'Z_CostContact',
                      'Z_Revenue'])

last_date = date(2014,7, 1)
df['Dt_enroll'] = pd.to_datetime(df['Dt_enroll'],format = '%d-%m-%Y')
df['Dt_enroll'] = pd.to_numeric(df['Dt_enroll'].dt.date.apply(lambda x: (last_date - x)).dt.days, downcast='integer')

print(df.columns)
df
```
